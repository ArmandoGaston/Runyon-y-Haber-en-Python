## Ejercicios 21

```python
import pandas as pd

datos_ej21 = pd.DataFrame({
    'Área académica' : ['Administración de empresas', 'Enseñanza', 'Humanidades', 'Ciencias', 'Ciencias sociales'],
    'Hombres': [400, 50, 150, 250, 200],
    'Mujeres': [100, 150, 200, 100, 200]
})

datos_ej21
```

Para hacer los cáculos más fácilmente utilizando las herramientas de Pandas, agregamos una nueva columna a la matriz anterior en la que calculamos los totales por filas.

Usamos el método *sum()*, cuyo argumento *axis = 1* permite sumar por filas

```python
datos_ej21['Total'] = datos_ej21[['Hombres', 'Mujeres']].sum(axis = 1)
datos_ej21
```

```python
mujeres_xaa = [round(x*100,2) for x in (datos_ej21['Mujeres']/datos_ej21['Total'])]

datos_ej21['% Mujeres'] = mujeres_xaa # Lo agregamos al df para una mejor presentación de resultados
datos_ej21
```

```python
total_hombres = datos_ej21['Hombres'].sum()

p_hombres = [round((x/total_hombres)*100,2) for x in datos_ej21['Hombres']]
```

Esta vez, en lugar de añadir la lista al df, lo que hacemos es utilizar *zip()* para emparejar cada ítem del área académica con su respectivo porcentaje.

```python
for n, x in zip(datos_ej21['Área académica'], p_hombres):
    print(f"{n}: {x}%")
```

```python
total_mujeres = datos_ej21['Mujeres'].sum()

p_mujeres = [round((x/total_mujeres)*100,2) for x in datos_ej21['Mujeres']]
for n, x in zip(datos_ej21['Área académica'], p_mujeres):
    print(f"{n}: {x}%")
```

```python
porcentaje_mujeres_aa = round((datos_ej21['Mujeres'].sum()/(datos_ej21['Mujeres'].sum() + datos_ej21['Hombres'].sum())*100), 2)
porcentaje_mujeres_aa
```

```python
porcentaje_hombres_aa = 100 - porcentaje_mujeres_aa
porcentaje_hombres_aa
```
