## Cálculo en Stats

```python
grupo1 = [91, 46, 108, 99, 110, 105, 191, 57, 34, 81]
grupo2 = [81, 51, 63, 51, 46, 45, 66, 64, 90, 28]
```

### Prueba U de Mann-Whitney

```python
from scipy.stats import mannwhitneyu

# Prueba de Mann-Whitney U (bilateral por defecto en versiones recientes)
stat, p = mannwhitneyu(grupo1, grupo2, alternative='two-sided')

print(f'Estadístico U: {stat:.4f}')
print(f'Valor-p: {p:.4f}')
```

### Prueba de Wilcoxon

La función *wilcoxon()* devuelve el valor absoluto del estadístico T, sin considerar el signo

```python
from scipy.stats import wilcoxon

stat, p = wilcoxon(grupo1, grupo2, alternative='two-sided')

print(f'Estadístico T: {stat:.4f}')
print(f'Valor-p: {p:.4f}')
```

```python
from scipy.stats import binomtest

# Calcular diferencias
diferencias = [a - b for a, b in zip(grupo1, grupo2)]
# Zip parea cada elemento del grupo 1 (a) con el correspondiente elemento del grupo 2 (2) por índice


# Contar signos
positivos = sum([1 for d in diferencias if d > 0])
negativos = sum([1 for d in diferencias if d < 0])
empates = sum([1 for d in diferencias if d == 0])

# Número de pares sin empates
n = positivos + negativos

# Prueba binomial (H0: p = 0.5)
test = binomtest(positivos, n=n, p=0.5, alternative='two-sided')

# Resultados
print(f"Diferencias: {diferencias}")
print(f"Positivos: {positivos}, Negativos: {negativos}, Empates: {empates}")
print(f"Estadístico binomial: {positivos} de {n}")
print(f"Valor-p: {test.pvalue}")

# Decisión
alpha = 0.05
if test.pvalue < alpha:
    print("Rechazamos H0: hay diferencia significativa entre las muestras")
else:
    print("No se rechaza H0: no hay evidencia suficiente de diferencia")
```
