### Cáculo manual

La prueba binomial tiene la siguiente fórmula

image.png

La probabiliad de obtener k éxitos en una muestra de N ensayos de Bernoulli (experimento aleatorio **con solo dos resultados posibles**) es igual al producto entre el coeficiente binomial, la probabilidad de obtener un éxito en un ensayo (conocida de antemano) elevada a la cantidad de éxitos y el complemento de esta probabilidad elevado a la diferencia entre el número de ensayos y el número de éxitos.

A su vez, la fórmula del coeficiente binomial es:

image.png

```python
# Probabilides de éxito (P) y fracaso (Q) en un ensayo de Bernoulli. Esto debe conocerse con anterioridad
probabilidad_bernoulli_p = 0.5
probabilidad_bernoulli_q = 0.5

# Datos de ensayos y éxitos
ensayos = 9

exitos0 = 0
exitos1 = 1
exitos2 = 2
exitos3 = 3
exitos4 = 4
exitos5 = 5
exitos6 = 6
exitos7 = 7
exitos8 = 8
exitos9 = 9

# Usar la fórmula para determinar la probabilidad de cada cantidad de éxitos en particular
# Usamos math.factorial para encontrar los factoriales del coeficiente binomial

coeficiente_binomial0 = math.factorial(ensayos) / (math.factorial(exitos0) * math.factorial(ensayos - exitos0))
coeficiente_binomial1 = math.factorial(ensayos) / (math.factorial(exitos1) * math.factorial(ensayos - exitos1))
coeficiente_binomial2 = math.factorial(ensayos) / (math.factorial(exitos2) * math.factorial(ensayos - exitos2))
coeficiente_binomial3 = math.factorial(ensayos) / (math.factorial(exitos3) * math.factorial(ensayos - exitos3))
coeficiente_binomial4 = math.factorial(ensayos) / (math.factorial(exitos4) * math.factorial(ensayos - exitos4))
coeficiente_binomial5 = math.factorial(ensayos) / (math.factorial(exitos5) * math.factorial(ensayos - exitos5))
coeficiente_binomial6 = math.factorial(ensayos) / (math.factorial(exitos6) * math.factorial(ensayos - exitos6))
coeficiente_binomial7 = math.factorial(ensayos) / (math.factorial(exitos7) * math.factorial(ensayos - exitos7))
coeficiente_binomial8 = math.factorial(ensayos) / (math.factorial(exitos8) * math.factorial(ensayos - exitos8))
coeficiente_binomial9 = math.factorial(ensayos) / (math.factorial(exitos9) * math.factorial(ensayos - exitos9))

probabilidad_exitos0 = coeficiente_binomial0 * (probabilidad_bernoulli_p**exitos0) * (probabilidad_bernoulli_q**(ensayos - exitos0))
probabilidad_exitos1 = coeficiente_binomial1 * (probabilidad_bernoulli_p**exitos1) * (probabilidad_bernoulli_q**(ensayos - exitos1))
probabilidad_exitos2 = coeficiente_binomial2 * (probabilidad_bernoulli_p**exitos2) * (probabilidad_bernoulli_q**(ensayos - exitos2))
probabilidad_exitos3 = coeficiente_binomial3 * (probabilidad_bernoulli_p**exitos3) * (probabilidad_bernoulli_q**(ensayos - exitos3))
probabilidad_exitos4 = coeficiente_binomial4 * (probabilidad_bernoulli_p**exitos4) * (probabilidad_bernoulli_q**(ensayos - exitos4))
probabilidad_exitos5 = coeficiente_binomial5 * (probabilidad_bernoulli_p**exitos5) * (probabilidad_bernoulli_q**(ensayos - exitos5))
probabilidad_exitos6 = coeficiente_binomial6 * (probabilidad_bernoulli_p**exitos6) * (probabilidad_bernoulli_q**(ensayos - exitos6))
probabilidad_exitos7 = coeficiente_binomial7 * (probabilidad_bernoulli_p**exitos7) * (probabilidad_bernoulli_q**(ensayos - exitos7))
probabilidad_exitos8 = coeficiente_binomial8 * (probabilidad_bernoulli_p**exitos8) * (probabilidad_bernoulli_q**(ensayos - exitos8))
probabilidad_exitos9 = coeficiente_binomial9 * (probabilidad_bernoulli_p**exitos9) * (probabilidad_bernoulli_q**(ensayos - exitos9))

# Mostrar cada probabilidad
tabla_prueba_binomial = pd.DataFrame({
    'Cantidad de Éxitos': [exitos0, exitos1, exitos2, exitos3, exitos4, exitos5, exitos6, exitos7, exitos8, exitos9],
    'Probabilidad': [probabilidad_exitos0, probabilidad_exitos1, probabilidad_exitos2, probabilidad_exitos3, probabilidad_exitos4, probabilidad_exitos5, probabilidad_exitos6, probabilidad_exitos7, probabilidad_exitos8, probabilidad_exitos9]
})
tabla_prueba_binomial
```

## Ejercicio 5

La serie Mundial puede durar de cuatro a siete juegos. Durante el período 1922-1979 la distribución del número de partidos fue la que se muestra a continuación. Examinar la hipótesis de que cada número de partidos tiene igual probabilidad de ocurrir.

```python
datos_ej5 = pd.DataFrame({
    'Número de partidos': [4, 5, 6, 7],
    'Frecuencia de realización': [11, 11, 11, 25]
})

datos_ej5
```

```python
from scipy.stats import chi2

suma = sum(datos_ej5['Frecuencia de realización'])
frecuencia_esperada = suma / 4

chi_cuadrado = sum([(x-frecuencia_esperada)**2/frecuencia_esperada for x in datos_ej5['Frecuencia de realización']])
p_valor = 1-chi2.cdf(chi_cuadrado, df = 3)

print("Chi cuadrado: ", round(chi_cuadrado, 4))
print("p-valor: ", round(p_valor, 4))
```

```python
from scipy.stats import chisquare

chi_cuadrado = chisquare(f_obs = datos_ej5['Frecuencia de realización'], f_exp = frecuencia_esperada)
chi_cuadrado
```

## Ejercicio 8

Construir la distribución muestral de la binomial cuando P = 0.20 Q = 0.80 y N = 6

### Cálculo manual

```python
n = 6
p = 0.20
q = 0.80

probabilidades = []
for x in range (0, 7):
  coeficiente_binomial = math.factorial(n) / (math.factorial(x) * math.factorial(n-x))
  probabilidad = coeficiente_binomial * (p**x) * (q**(n-x))
  probabilidades.append(probabilidad)

distribucion_ej8 = pd.DataFrame({'x': range(0, 7), 'P(x)': probabilidades})
distribucion_ej8
```

## Ejercicio 13

```python
#from scipy.stats import binomtest

# Datos
n = 46          # Tamaño de muestra
x = 28          # Éxitos (a favor)
p0 = 0.5        # Proporción bajo H0
alpha = 0.05    # Nivel de significancia

# Prueba binomial exacta (bilateral)
test = binomtest(k=x, n=n, p=p0, alternative='two-sided')

# Resultados
print(f'Valor-p: {test.pvalue:.4f}')
print(f'Estadístico observado: {x} de {n}')

# Decisión
if test.pvalue < alpha:
    print("Rechazamos H0: hay evidencia de que la proporción no es 0.5")
else:
    print("No se rechaza H0: no hay evidencia suficiente para decir que la proporción difiere de 0.5")
```
