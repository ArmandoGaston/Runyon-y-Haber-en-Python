## Ejercicio 15

```python
import statistics
import scipy.stats as stats
from scipy.stats import norm

x = 100
s = 16

dist_norm = statistics.NormalDist(x, s)

inciso_a = 1 - dist_norm.cdf(122)
inciso_b = norm.ppf(0.95, loc = x, scale = s)
inciso_c = [norm.ppf(0.975, loc = x, scale = s), norm.ppf(0.025, loc = x, scale = s)]
inciso_d = norm.cdf(100, loc = x, scale = s) - norm.cdf(90, loc = x, scale = s)
inciso_ei = 1- norm.cdf(122, x, s)
inciso_eii = norm.cdf(110, x, s) - norm.cdf(90, x, s)
inciso_eiii = inciso_ei * inciso_eii

print(f'Inciso a: {round(inciso_a, 4)}')
print(f'Inciso b: {round(inciso_b, 4)}')
print(f'Inciso c: \n  Valor superior: {round(inciso_c[0], 2)}\n  Valor inferior {round(inciso_c[1], 2)}')
print(f'Inciso d: {round(inciso_d, 2)}')
print(f'Inciso e: \n  i: {round(inciso_ei, 2)}\n  ii: {round(inciso_eii, 2)}\n  iii: {round(inciso_eii, 2)}')
```

## Ejercicio 20

```python
x = 0.01
s = 0.001

dist_norm = statistics.NormalDist(x, s)

probabilidad_defectuoso = 1 - dist_norm.cdf(x + 2*s) + dist_norm.cdf(x - 2*s)
probabilidad_defectuoso_2 = probabilidad_defectuoso ** 2
probabilidad_no_defectuoso = 1 - probabilidad_defectuoso
probabilidad_no_defectuoso2 = probabilidad_no_defectuoso ** 2
probabilidad_largo = probabilidad_defectuoso/2
probabilidad_corto2 = (probabilidad_no_defectuoso/2)**2

print(f'Probabilidad de que el producto sea defectuoso: {round(probabilidad_defectuoso, 4)}')
print(f'Probabilidad de que dos productos sean defectuosos: {round(probabilidad_defectuoso_2, 4)}')
print(f'Probabilidad de que el producto no sea defectuoso: {round(probabilidad_no_defectuoso, 4)}')
print(f'Probabilidad de que dos productos no sean defectuosos: {round(probabilidad_no_defectuoso2, 4)}')
print(f'Probabilidad de que el producto sea demasiado largo: {round(probabilidad_largo, 4)}')
print(f'Probabilidad de que el dos productos sean demasiado cortos: {round(probabilidad_corto2, 4)}')
```
