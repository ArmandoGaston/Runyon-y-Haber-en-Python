### Realización manual

Cálculo de q crítico

```python
from scipy.stats import distributions

# Parámetros
k = 4  # Número de grupos
df_within = 36  # Grados de libertad dentro de los grupos
alpha = 0.01  # Nivel de significancia

# Calcular el valor crítico q
q_critico = distributions.studentized_range.ppf(1 - alpha, k, df_within)

print(f"Valor crítico de q: {q_critico:.4f}")
```

Suma de cuadrados dentro de los grupos (*within*)

```python
datos_ej1['Invierno**2'] = datos_ej1['Invierno']**2
datos_ej1['Primavera**2'] = datos_ej1['Primavera']**2
datos_ej1['Verano**2'] = datos_ej1['Verano']**2
datos_ej1['Otoño**2'] = datos_ej1['Otoño']**2
datos_ej1
```

```python
suma1 = (datos_ej1['Invierno'].sum())**2
suma2 = (datos_ej1['Primavera'].sum())**2
suma3 = (datos_ej1['Verano'].sum())**2
suma4 = (datos_ej1['Otoño'].sum())**2

print('Suma de Invierno al cuadrado:', round(suma1, 2))
print('Suma de Primavera al cuadrado:', round(suma2, 2))
print('Suma de Verano al cuadrado:', round(suma3, 2))
print('Suma de Otoño al cuadrado:', round(suma4, 2))
```

```python
sc1 = datos_ej1['Invierno**2'].sum() - suma1/10
sc2 = datos_ej1['Primavera**2'].sum() - suma2/10
sc3 = datos_ej1['Verano**2'].sum() - suma3/10
sc4 = datos_ej1['Otoño**2'].sum() - suma4/10
suma_cuadrados_within = sc1 + sc2 + sc3 + sc4
round(suma_cuadrados_within, 3)
```

```python
varianza_within = suma_cuadrados_within / 36
varianza_within
```

```python
dhs = q_critico * np.sqrt(varianza_within/ 10)
dhs
```

Para hacer las comparaciones usamos la función *np.where*, que es equivalente a una función *Si* en Excel:

- Condición

- Qué hacer si la condición se cumple para el caso (valor si verdadero).

- Qué hacer si la condición *no* se cumple (valor si falso.)

```python
comparacion = pd.DataFrame()
comparacion['Meses'] = ['Invierno - Primavera', 'Invierno - Verano', 'Invierno - Otoño', 'Primavera - Verano','Primavera - Otoño', 'Verano - Otoño']
comparacion['Módulo de diferencias'] = [abs(np.mean(datos_ej1['Invierno']) - np.mean(datos_ej1['Primavera'])),
                                        abs(np.mean(datos_ej1['Invierno']) - np.mean(datos_ej1['Verano'])),
                                        abs(np.mean(datos_ej1['Invierno']) - np.mean(datos_ej1['Otoño'])),
                                        abs(np.mean(datos_ej1['Primavera']) - np.mean(datos_ej1['Verano'])),
                                        abs(np.mean(datos_ej1['Primavera']) - np.mean(datos_ej1['Otoño'])),
                                        abs(np.mean(datos_ej1['Verano']) - np.mean(datos_ej1['Otoño']))]

comparacion['Decision sobre H0'] = np.where(
    comparacion['Módulo de diferencias'] >= dhs, 'Rechazo', 'No rechazo'
)
comparacion
```

## Ejercicio 9

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

nueva_inglaterra = [8, 7, 8, 6.5, 8, 8]
medio_este = [7, 7, 8, 7, 7, 7]
lejano_oeste = [9, 7, 6, 7, 8, 5]

datos_ej9 = pd.DataFrame({
            'Regiones': ['Nueva Inglaterra'] * 6 + ['Medio Este'] * 6 + ['Lejano Oeste'] * 6,
            'Valores': nueva_inglaterra + medio_este + lejano_oeste
})

datos_ej9
```

```python
# Modelo de ANOVA
modelo = ols('Valores ~ Regiones', data= datos_ej9).fit()
anova_resultados = anova_lm(modelo)

# Mostrar resultados
print(anova_resultados)
```
