### Parte A
Tendría que ser la media, porque la suma de los desvíos es igual a 0

```python
resultado_14a = round((150 * 3 + 400 * 2.50 + 100 * 4.25 + 200 * 3.5)/(150 + 400 + 100 + 200), 2)
resultado_14a
```

Noten que es una media ponderada en la que los pesos son las cantidades compradas.

### Parte B

Con la ayuda del libro, vemos que sencillamente incorporamos los costos de las comisiones al cálculo del numerador y ya tenemos la respuesta. 

Sin embargo, para entender por qué se hace esto, se realiza un segundo procedimiento. Uno puede sumar al costo total de las operaciones bursátiles de cada tipo de acción su correspondiente costo de comisiones, luego dividir por la cantidad que se compró, y se tiene el precio efectivo por unidad. 

Es decir, que si yo compré 2 acciones a $2 en total, el precio por unidad es obviamente $1. Pero si además pagué una comisión total de $0.50, entonces el precio es ($2 + 0.5)/2 = $1.25.

Al tener los nuevos precios, se los reemplaza en el cálculo. El resultado es el mismo. 

```python
resultado_14b = round((13.5 + 30 + 12.75 + 21 + 150 * 3 + 400 * 2.50 + 100 * 4.25 + 200 * 3.5)/(150 + 400 + 100 + 200), 2)
print('Nuevo precio de equilibrio = ',resultado_14b)

precio_a = (150 * 3 + 13.5)/150
precio_b = (400 * 2.5 + 30)/400
precio_c = (100 * 4.25 + 12.75)/100
precio_d = (200 * 3.5 + 21)/200

print(f"""{'-'*35}
Precio a: {precio_a}
{'-'*35}
Precio b: {precio_b}
{'-'*35}
Precio c: {precio_c}
{'-'*35}
Precio d: {precio_d}
{'-'*35}""")

resultado_14b = round((150 * precio_a + 400 * precio_b + 100 * precio_c + 200 * precio_d)/(150 + 400 + 100 + 200),2)
print('Nuevo precio de equilibrio = ',resultado_14b)         
```

```python
data = {"REGION": ['Norteamérica', 'Sudamérica', 'Europa', 'Rusia', 'África', 'Asia'],
        """POBLACIÓN ACTUAL (EN MILLONES)""": [225,276,456,241,344,1990],
        "TASA ANUAL DE CRECIMIENTO (PORCENTAJE)": [1.1,2.9,0.8,1.0,2.4,2.0]}

df= pd.DataFrame(data)
df

```

```python
datos_ej35 = [n/2 for n in datos_ej30_cap3]

distribucion_frecuencia35 =pd.DataFrame(
    {'Media': pd.Series(datos_ej35).value_counts().sort_index().index,
    'Frecuencia' : (pd.Series(datos_ej35).value_counts().sort_index().values)})
distribucion_frecuencia35['Frecuencia relativa'] = distribucion_frecuencia35['Frecuencia']/distribucion_frecuencia35['Frecuencia'].sum()
distribucion_frecuencia35['Frecuencia relativa acumulada'] = distribucion_frecuencia35['Frecuencia relativa'].cumsum()
distribucion_frecuencia35
```

```python
sns.barplot(x = distribucion_frecuencia35['Media'], y = distribucion_frecuencia35['Frecuencia'], edgecolor = 'black', alpha = 0.8)

plt.title('Media de lanzamiento de dos dados')
plt.ylabel("Frecuencia")
plt.show()
```
